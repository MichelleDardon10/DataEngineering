```python

tidy_df = pd.melt(
    df,
    id_vars=['country', 'measurement_unit', 'notes'],
    var_name='year',
    value_name='value'
)


tidy_df = tidy_df.dropna(subset=['value'])


tidy_df = tidy_df[['country', 'year', 'measurement_unit', 'value', 'notes']]

tidy_df.head()
```

```python

flights_df = pd.read_csv("/Users/michelledardon/Documents/02_DE/flights.csv")

pivot_df = flights_df.pivot(index='year', columns='month', values='passengers')


month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]
pivot_df = pivot_df[month_order]

pivot_df.head()
# Show the result
print(pivot_df)
```
